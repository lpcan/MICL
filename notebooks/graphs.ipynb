```python
from augmentations import *

def augmenter(input_shape, mean=0.948, std=1.108):
    return keras.Sequential(
        [
            layers.Input(shape=input_shape),
            # layers.Normalization(mean=mean, variance=std**2),
            layers.RandomFlip(mode='horizontal_and_vertical'),
            RandomGaussianNoise(stddev=0.017359),
            RandomRotate90(),
        ]
    )

aug = augmenter((224,224,1))
img = np.clip(cutout, a_min=0, a_max=10)
img = np.arcsinh(img / 0.017359)
img = np.expand_dims(img, 0)
img = np.expand_dims(img, -1)
new_img = aug(img)
plt.imshow(new_img.numpy().squeeze())
plt.xticks([])
plt.yticks([])
plt.show()
```

```python
from measure_sb_cut.scripts import measure_icl
from data.scripts.display_cutouts import stretch
import h5py
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import ascii
from astropy.cosmology import FlatLambdaCDM
from astropy.convolution import Gaussian2DKernel, convolve
from astropy.table import join
from scipy import spatial

num = '51'
cutouts = h5py.File('/srv/scratch/z5214005/cutouts_550.hdf')
tbl = ascii.read('/srv/scratch/z5214005/camira_final.tbl')
zs = tbl['z_cl']

members = ascii.read('/srv/scratch/z5214005/camira_s20a_dud_member.dat', 
                    names=['RA_cl', 'Dec_cl', 'Richness', 'z_cl', 'RA', 'Dec', 'M', 'w'])

# Match this catalogue to the cluster catalogue
merged = join(members, tbl, keys_left=['RA_cl', 'Dec_cl'], keys_right=['RA [deg]', 'Dec [deg]'])
merged = merged['ID', 'Name', 'RA_cl', 'Dec_cl', 'z_cl_1', 'RA', 'Dec']

cosmo = FlatLambdaCDM(H0=68.4, Om0=0.301)
cutout = cutouts[num]['HDU0']['DATA']
BAD = 1
BRIGHT_OBJECT = 512
mask_data = np.array(cutouts[num]['HDU1']['DATA']).astype(int)
bad_mask = np.array(mask_data & (BAD | BRIGHT_OBJECT)).astype(bool)

# Background estimate
bkg = measure_icl.background_estimate(cutout, zs[int(num)], cosmo, mask=bad_mask)
bkg_subtracted = cutout - bkg

# Segment the image
cold_labels = measure_icl.create_cold_labels(cutout, bad_mask)
cold_mask = measure_icl.enlarge_mask(cold_labels, sigma=2)

# Unsharp mask the image for hot mask creation
kernel = Gaussian2DKernel(5)
conv_img = convolve(np.array(cutout), kernel)
unsharp = np.array(cutout) - conv_img

# Create the hot mask
hot_mask_bkg = measure_icl.background_estimate(unsharp, z=zs[int(num)], cosmo=cosmo, mask=bad_mask)
hot_labels = measure_icl.create_hot_labels(unsharp, (bad_mask + cold_mask), background=hot_mask_bkg)
hot_mask = measure_icl.enlarge_mask(hot_labels, sigma=1)

# Mark the cluster members in the cold mask
x_locs, y_locs = measure_icl.get_member_locs(int(num), merged, cutout.shape)
c_members = cold_labels[y_locs.astype(int), x_locs.astype(int)]

# Mask out non cluster members
member_mask = np.isin(cold_labels, c_members) | (cold_labels == 0)
non_member_mask = ~member_mask
non_member_mask = measure_icl.enlarge_mask(non_member_mask, sigma=2)
non_member_mask = non_member_mask + hot_mask
# Member mask keeps the background and cluster members
member_mask = ~non_member_mask

# Make the circular mask
mid = (cutout.shape[0] // 2, cutout.shape[1] // 2)
bcg_label = cold_labels[mid[0], mid[1]]

pts = np.array(np.argwhere(cold_labels == bcg_label))

# Find points that are furthest apart
candidates = pts[spatial.ConvexHull(pts).vertices]
dist_mat = spatial.distance_matrix(candidates, candidates)
i, j = np.unravel_index(dist_mat.argmax(), dist_mat.shape)
pt1 = candidates[i]
pt2 = candidates[j]

size = np.sqrt((pt1[0] - pt2[0])**2 + (pt1[1] - pt2[1])**2)
radius = np.max((size, cosmo.arcsec_per_kpc_proper(zs[int(num)]).value * 100 * 1/0.168))
l = mid[0] - radius
h = mid[0] + radius

plt.imshow(stretch(cutout) * ~bad_mask)
plt.xlim(l, h)
plt.ylim(l, h)
plt.show()

plt.imshow(stretch(bkg_subtracted) * ~bad_mask)
plt.xlim(l, h)
plt.ylim(l, h)
plt.show()

plt.imshow(stretch(cutout) * member_mask)
plt.xlim(l, h)
plt.ylim(l, h)
plt.show()
```
